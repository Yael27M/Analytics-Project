```python
import requests
import json
import pandas as pd
from groq import Groq
import re

# -----------------------------
# 1. INIT LLM CLIENT
# -----------------------------
client = Groq(api_key="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")

# -----------------------------
# 2. EBAY API CALL
# -----------------------------
token = "v^1.1#i^1#p^3#I^3#r^0#f^0#t^H4sIAAAAAAAA/+Vaa2xbVx2P82jVbS3TBjSqJmS87Uuja9+H77V9iY2cxknc5mmnaRvYsnPPPdc+zX1495ybxNFU0qBGGtOmthoIBNPKh1WV4MMQ7APbOrFWKmiso+IhdWiIhVYaErAhmISENsa5TuK6gbZxXJgR94t9zvm/fv/z///P415+Ycu23UsDS3/bHtjaemqBX2gNBIQ7+W1bOrp2tLXu6mjhawgCpxYeWGhfbPt9NwGWWVJziJQcm6DgnGXaRK10JkOea6sOIJioNrAQUSlU8+mhQVUM82rJdagDHTMUzPYmQzLUeUFEksjzKB6TdNZrr8kcd5IhUVB4RZOkuBaFIhANNk6Ih7I2ocCmbJwXFY6XOVEaF0RVjqo8H07EpclQcAK5BDs2IwnzoVTFXLXC69bYenNTASHIpUxIKJVN9+VH0tnezPB4d6RGVmrVD3kKqEeub+1xdBScAKaHbq6GVKjVvAchIiQUSa1ouF6oml4zZhPmV1wdVaCWMPioEoeanJBit8WVfY5rAXpzO/werHNGhVRFNsW0fCuPMm9ohxGkq61hJiLbG/R/xjxgYgMjNxnK9KQP7c9ncqFgfnTUdWawjnQfqRiPxuNRKSbKoRS2OFmJiqsqVuSsOnidjj2OrWPfXSQ47NAexOxF670SrfEKIxqxR9y0QX1baumia96LRSf96VyZP48WbX9GkcVcEKw0b+37tWC4Nv23Kxx0OcZiQdcEBKKyrMg3CAc/1+sKiZQ/K+nR0YhvC9JAmbOAO41oyQQQcZC517OQi3VVkg1RihuI05WEwUUThsFpsq5wgoEQj5CmwUT8/yMyKHWx5lFUjY71AxV4yVAeOiU06pgYlkPrSSp1ZjUW5kgyVKS0pEYis7Oz4Vkp7LiFCCuxQuTg0GAeFpEFQlVafGtiDleiAiLGRbBKyyVmzRwLOqbcLoRSkquPApeW88g0WcdayF5nW2p97w1A7jEx88A4U9FcGAccQpHeEDQdzWCIprDeXMhEMR6XYnFF9HM9wfNiQyBNp4DtIUSLTpPB7B8Z6R/MNISNlU9AmwtVTXXh5bUqJEkcH2P/GgKbLpWyluVRoJko22RzKcf5mJJoCF7J85otEWdFaJdn0Bw1SUPQ/FVXxcBQqTON7H9fSv1c/zix5jJ9uUx+YGp8ZF9muCG0OWS4iBTHfazNFqfpsfRgmj1D/TGnOHZYAdH+WI+ecw7PjJmH+vZqpPToxFiP7LhQmssOKHMo3aOMFQ72R3omRgWcGDkoI9grztj708lkQ07KI+iiJitdfUZ+RN8LvMOT2YlYD+rar/WPEsXtnRwe6DJpYV9aOwSFKO0x42ONgR8qNFumV5fchpfb8RuneBWgn+sfA0h3JTGnKlVoirUaApopNF29RnFFEeUEEhI6DxQxJsR5IInsIMMePaErDS+/TYY3awFg54uMzv/DjeZ6OQ2guA6hbnCiwFwgS1qDi3KzzfHtWpOJf3b7L0Dzc70OeL4MwoSAEg7724YwdKyIAzxa9LumKlYHN0IUIezsF1456jPJYRcB3bHN8maY6+DB9gw7LTpueTMKq8x18AAIHc+mm1G3yloHh+GZBjZN/0pgMwpr2Osx0wZmmWJINqUS2360kTpYSqBcAahjUvLzZUOcrM9CLkRhrK9cKG7GWBcxhaByibYZpjpVVk22HYoNDFdkEE8j0MWljVtxTY6f6zeStRl/EJYLdU3dCsOGVNVwIR2ZeAZtNO2qeBmLs6nSYIFSacNlparOQoSAQr3xaCCkawBO18lGirhiY2P3S46FITab7OKst8E7F6RjF0E65bm4uYBVtkNTbD9UxDq3bm/EATT/KGbh0xB2P1pWr9L8XG8u9L234ezei2aabZMbldjZTNQhBw1B56JRLcFpkgi5WCIBhURMERQoNIS56a4QhVgsIUf5uNTYhVoOAdNqLmQl19E96C+H/1PI2o8e/4+AW9dR87rpX94xRq5/vZ9qqTzCYuAcvxh4pTUQ4Lv5B4X7+c9uadvf3nbXLoIp24sBI0xwwQbUc1F4GpVLALut97a89qs3hz/z0t4zj1/duXDsgcjJlh01XxeceojvrH5fsK1NuLPmYwP+vmsjHcIndm4XFV4WJUFk4cpP8vdfG20XPt3+SXpgede5r4FfHntWeMM68/iSePl0J7+9ShQIdLS0LwZaJh87luxqf/pzb3+v46TQ2Ro8svUHhcJTr89dmnzxD5b60F8WX0x967l7ls9dfIsMLr9wqOtLdOsd08cf+fE9ub6OILy7/83L9118WPnCO689ta9d/+EHJ379xDsf/ukF9XdnXj82P/H87k7zu18+svjBpece+433lct9sZ3vwp++f/LkdxZP7zieXkq/tTXZdteej96/euXlbyxfevLhnyu5+N+j5xfOfv7K/M86X33vjsK9r/71w/Yjfzaf7L6Qn+958ITuXHlGeK+87RH+t1+84NLQt5Wz//hq4eJS5uwrU5964keFl//4zTdaD0SePzF99KMLp7sXTxzYvfyLC87X+17iMs8Gz/8k8O7bxaHwmauXj35//nx4enJlLv8Jb1yJEvchAAA="
url = "https://api.ebay.com/buy/browse/v1/item_summary/search"

headers = {
    "Authorization": f"Bearer {token}",
    "X-EBAY-C-MARKETPLACE-ID": "EBAY_US"
}

params = {
    "q": ["Rolex 2030 4429","Rolex 1570 7831","Rolex 3030 5309", "Rolex 4030 502", "Rolex 4030 311", "Tudor 390 260"]
    #"limit": 20
}

response = requests.get(url, headers=headers, params=params)
data = response.json()

listings = data.get("itemSummaries", [])


# -----------------------------
# 3. GET ITEM DETAILS
# -----------------------------
def get_item_details(item_id):
    url = f"https://api.ebay.com/buy/browse/v1/item/{item_id}"
    r = requests.get(url, headers=headers)
    return r.json() if r.status_code == 200 else {}


#build full dataset (IMPORTANT: multiple rows)
final_data = []

for item in listings:
    item_id = item.get("itemId")
    if item_id:
        final_data.append(get_item_details(item_id))


# -----------------------------
# 4. LLM EXTRACTION
# -----------------------------
def extract_with_llm(title,condition):

    # -----------------------------
    # RULE-BASED EXTRACTION FIRST
    # -----------------------------

    title_lower = title.lower()
    condition = condition.lower()

    # brand
    brand = "Rolex" if "rolex" in title_lower else None

    # component type
    components = [
        "escape wheel",
        "balance wheel",
        "third wheel",
        "rotor",
        "stem",
        "barrel",
        "barrel arbor",
        "mainspring"
    ]

    component_type = None
    for comp in components:
        if comp in title_lower:
            component_type = comp
            break

    # #condition
    if condition in ("new with tags", "new without tags"):
        condition = "New"
    elif condition in ("pre-owned"):
        condition = "Used"
    elif condition in ("refurbished"):
        condition = "Refurbished"
    else:
      condition = None

    # -----------------------------
    # NUMERIC EXTRACTION
    # -----------------------------

    numbers = re.findall(r'\d+', title)

    calibers = []
    part_numbers = []

    for num in numbers:
        if len(num) == 4:
            n = int(num)

            # Rolex calibers usually lower range
            if 1000 <= n <= 3999:
                calibers.append(num)

            # Part numbers usually higher
            if 4000 <= n <= 9999:
                part_numbers.append(num)

    caliber = calibers[0] if calibers else None
    part_number = part_numbers[0] if part_numbers else None

    # -----------------------------
    # FALLBACK TO LLM ONLY IF MISSING
    # -----------------------------

    if brand and caliber and part_number and component_type:
        return {
            "brand": brand,
            "caliber": caliber,
            "part_number": part_number,
            "condition": condition,
            "component_type": component_type
        }

    prompt = f"""
    Extract:

    brand
    caliber
    part_number
    condition
    component_type

    Return only JSON.

    Title:
    {title}
    """

    try:
        completion = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        output = completion.choices[0].message.content
        llm_result = json.loads(output)

        return {
            "brand": brand or llm_result.get("brand"),
            "caliber": caliber or llm_result.get("caliber"),
            "part_number": part_number or llm_result.get("part_number"),
            "condition": condition or llm_result.get("condition"),
            "component_type": component_type or llm_result.get("component_type")
        }

    except:
        return {
            "brand": brand,
            "caliber": caliber,
            "part_number": part_number,
            "condition": condition,
            "component_type": component_type
        }


# -----------------------------
# 5. PIPELINE
# -----------------------------
def extract_ebay_data(final_data):

    rows = []

    for item in final_data:

        title = item.get("title")
        condition = item.get("condition")

        #price
        price = item.get("price", {})
        price_value = float(price.get("value", 0)) if price.get("value") else 0
        currency = price.get("currency")

        #shipping
        shipping_options = item.get("shippingOptions", [])
        shipping_costs = []

        for s in shipping_options:
            cost = s.get("shippingCost", {}).get("value")
            if cost:
                shipping_costs.append(float(cost))

        shipping_cost = min(shipping_costs) if shipping_costs else 0
        net_price = price_value + shipping_cost

        #seller
        seller = item.get("seller", {})
        seller_score = seller.get("feedbackScore")
        seller_pct = seller.get("feedbackPercentage")

        #availability
        availability = item.get("estimatedAvailabilities", [{}])[0]
        sold_qty = availability.get("estimatedSoldQuantity")
        remaining_qty = availability.get("estimatedRemainingQuantity")

        #location
        location = item.get("itemLocation", {})
        country = location.get("country")

        #time
        created = item.get("itemCreationDate")

        # ---------------- LLM ----------------
        extracted = extract_with_llm(title or "", condition)

        #synthetic id
        synthetic_id = "_".join([
            extracted.get("brand") or "UNK",
            extracted.get("caliber") or "UNK",
            extracted.get("part_number") or "UNK",
            extracted.get("condition") or "UNK",
            extracted.get("component_type") or "UNK"
        ])

        ##final append of data
        rows.append({
            "item_id": item.get("itemId"),
            "title": title,

            "price": price_value,
            "shipping_cost": shipping_cost,
            "net_price": net_price,

            "currency": currency,

            "brand": extracted.get("brand"),
            "caliber": extracted.get("caliber"),
            "part_number": extracted.get("part_number"),
            "condition": extracted.get("condition"),
            "component_type": extracted.get("component_type"),

            "synthetic_id": synthetic_id,

            "seller_score": seller_score,
            "seller_feedback_pct": seller_pct,

            "sold_qty": sold_qty,
            "remaining_qty": remaining_qty,

            "country": country,
            "created_at": created
        })

    return pd.DataFrame(rows)


# -----------------------------
# 6. MAIN
# -----------------------------
df = extract_ebay_data(final_data)

df.to_csv("watch_dataset.csv", index=False)

df
```
